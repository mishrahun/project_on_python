```python
import pandas as pd
import numpy as np
import matplotlib as pet
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\mishr\Downloads\Python_Diwali_Sales_Analysis-main\Python_Diwali_Sales_Analysis-main\Diwali_Sales_Data.csv",encoding='unicode_escape')
df
```

```python
df.rename(columns={'Age Group':'age_group'},inplace=True)
```

```python
sales_by_gender=df.groupby(['Gender'],as_index=False).sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_by_gender)
```

```python
females spends more than males in diwali
```

```python
sales_by_age=df.groupby(['age_group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
da=sns.barplot(x='age_group',y='Amount',data=sales_by_age)
```

```python
da=sns.barplot(x='age_group',y='Amount',data=sales_by_age)
for bar in da.containers:
    da.bar_label(bar)
```

```python
agw=sns.countplot(x='age_group',data=df,hue='Gender')
for bar in agw.containers:
    agw.bar_label(bar)
```

```python
specially(26-35) year old females spend more money as compared to man
```

```python
sales_by_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(16,5)})
x=sns.barplot(x='State',y='Amount',data=sales_by_state)
for bar in x.containers:
    x.bar_label(bar)
```

```python
sales_by_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(16,5)})
k=sns.barplot(x='State',y='Orders',data=sales_by_state)
for bar in k.containers:
    k.bar_label(bar)
```

```python
sales_by_marital=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

d=sns.barplot(x='Marital_Status',y='Amount',data=sales_by_marital,hue='Gender')
for bar in d.containers:
    d.bar_label(bar)
```

```python
sns.set(rc={'figure.figsize':(19,5)})
sns.countplot(x='Occupation',data=df)
```

```python
sales_by_occu=df.groupby(['Occupation'],as_index=False).sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Occupation',y='Amount',data=sales_by_occu)
```

```python
it sector, health care and aviation are the occupation who spends more.
```

```python
sns.set(rc={'figure.figsize':(27,5)})
sns.countplot(x='Product_Category',data=df)
```

```python
sales_by_occu=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
d=sns.barplot(x='Product_Category',y='Amount',data=sales_by_occu)
for bar in d.containers:
    d.bar_label(bar)
```

```python
people spend lot more on food as compared to clothings on diwali but food , clothing,electronics are the top spendings.
```

```python
most married women in age between 26-35 from uttar pradesh, maharashtra,and karanataka in department of  it,health,aviation spends lot on food,clothes and electronics in diwali
```
